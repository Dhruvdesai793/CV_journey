```python
import cv2

# Initialize webcam using 0
cap = cv2.VideoCapture(0)

cap.set(3, 640)
cap.set(4, 480)
# Set brightness (Property ID 10), not needed. my cam sucks
cap.set(10, 100) 

while True:
    success, img = cap.read()
    img = cv2.flip(img, 1) #inverts video horizontally
    if not success:
        print("Failed to capture image")
        break

    cv2.imshow("Webcam Feed", img)
    
    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release() # releases the webcam hardware
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

kernel = np.ones(8)

img = cv2.imread('luffy.png')

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(img, (9,9),0)
can = cv2.Canny(img, 200,150)
dilate = cv2.dilate(can, kernel, iterations = 2)
erode = cv2.erode(dilate, kernel, iterations= 3)

cv2.imshow('gray',gray)
cv2.imshow('blur',blur)
cv2.imshow('canny',can)
cv2.imshow('dilated',dilate)
cv2.imshow('eroded',erode)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = np.zeros((1012, 1412, 3), dtype="uint8")

# drawww
img[200:300, 100:900] = 255 
img[400:500, 200:300] = (234, 12, 234) 

cv2.line(img, (0, 0), (img.shape[1], img.shape[0]), (255, 255, 21), 3)
cv2.rectangle(img, (0, 0), (250, 340), (0, 232, 43), cv2.FILLED) 

cv2.circle(img, (400, 600), 300, (255, 0, 0), -1) 
cv2.putText(img, "OpenCV", (100, 900), cv2.FONT_HERSHEY_COMPLEX, 4, (0, 0, 255), 8)

# fixed niri conflicts!!
cv2.namedWindow("image", cv2.WINDOW_NORMAL) # window to be resized
cv2.resizeWindow("image", 600, 600)        

cv2.imshow("image", img)

# This loop keeps the window "alive" while waiting for a key
while True:
    if cv2.waitKey(1) & 0xFF == ord('q'): # Press 'q' to exit safely
        break
    # This check ensures the window hasn't been closed via the 'X' button
    if cv2.getWindowProperty("image", cv2.WND_PROP_VISIBLE) < 1:
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

luffy = cv2.imread('luffy.png')

zoro = cv2.putText(luffy, "Zoro",(710,660), cv2.FONT_HERSHEY_COMPLEX, 4, (0, 255, 0), 10)
sanji = cv2.putText(luffy, "Sanji",(1290,100), cv2.FONT_HERSHEY_COMPLEX, 4, (255, 0, 0), 10)
nami = cv2.putText(luffy, "Nami",(1150,300), cv2.FONT_HERSHEY_COMPLEX, 4, (0, 0, 255), 10)
usopp = cv2.putText(luffy, "Usopp",(300,250), cv2.FONT_HERSHEY_COMPLEX, 4, (255, 255, 0), 10)
luffy = cv2.putText(luffy, "Luffy",(500,900), cv2.FONT_HERSHEY_COMPLEX, 4, (255, 255, 255), 10)


cv2.rectangle(zoro, (743,340),(890,540 ), (0, 255, 0), 10) #zoro
cv2.rectangle(sanji, (1110,0),(1280,170 ), (255, 0, 0), 10) #sanji
cv2.ellipse(nami, (1040,250), (100, 130), 0, 0, 360, (0, 0, 255), 10) #nami
cv2.circle(usopp, (830, 200), 100, (255, 255, 0), 10) #usopp
cv2.polylines(luffy, [np.array([[500, 900], [600, 800], [700, 900]], dtype=np.int32)], isClosed=True, color=(255, 255, 255), thickness=10) #luffy

# 2. Create the window first and set it to NORMAL , as my Niri windows manager wasn't working well without it b4.
cv2.namedWindow("luffy", cv2.WINDOW_NORMAL) 

# 3. Resize the window 
cv2.resizeWindow("luffy", 600, 600)

cv2.imshow("luffy", luffy)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
